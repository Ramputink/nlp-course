# Importing modules

Python code can access to code from another Python module by importing the file/function using import. These modules can define functions, classes and/or variables.

Most Python distributions already include the most popular modules with predefined libraries which make our programmer lives easier. Some well-known libraries are: math, time, sys, os, numpy, etc.
    
There are several ways to import a library:

1. Import all the contents of the library: `import lib_name`.


````python
import math
print(math.pi)
````

````python
import time
print(time.time())  # returns the current processor time in seconds
time.sleep(2) # suspends execution for the given number of seconds
print(time.time()) # returns the current processor time in seconds again!!!
````

**Note:** When you want to use the functions, variables, ... of the imported modules, you have to call them as part of the library (`lib_name.method`).

2. Define a short name to use the library: ``import lib_name as lib``.

````python
import math as mt
print(mt.pi)
````

````python
import time as t
print(t.time())
````

3. Import only some elements of the library using the `from` statement.

````python
from math import pi
print(pi)
````

````python
from time import time, sleep
print(time())
sleep(2)
print(time())
````

**Note**: Note that in the above example, we used `math.pi` or `time.time`. Here we can directly use `pi` or `time`. Besides, when using the statement `"from"`, only the variable `pi` or the funtion `time` are imported, instead of importing the whole library.

However, we can import all the functions and variables as separate entities  using "*": `from lib_name import *`

````python
from math import *
print(pi)
print(factorial(6))
````

````python
from time import *
print(time())
sleep(2)
print(time())
````

### Pickle

Python pickles are the set of tools to convert any Python object, variable, ... as a chain of bytes (i.e., serialize any object into a byte stream). Once the object has been serialized, you can do many things with these bytes, such as storing them in a file or database. Later, you can "unpick" the byte string to recover the original Python object.

To use pickles in Python, we will use the Pickle module. This allows us to store almost any Python object, variable,... directly in a file or string without any additional operation.

Let's start by creating a pickle file from a Python list:

````python
mylist = ['apple', 'peach', 'orange', 'cherry']
````

In order to serialize `mylist` or to pickle it, we can use the function `.dumps()` of pickle:

````python
import pickle

mylistserialized = pickle.dumps(mylist)
print(mylistserialized)
````

You must serialize the data if you need to handle it as a data stream. The most common use is to save your information to a disk or transmit it over the network.
Data serialization encodes the data structure, variable names, variable values, and variable types, so that they can be reconstructed on the other side of the communication.

Now, we can recover the original list from the serialized object using using the funcion `.loads()`

````python
mylistrecovered = pickle.loads(mylistserialized)
print(mylistrecovered)
````

In case we want to directly save this object into a file to recover its value later, we can use the methods `.dump()` and `.load()` of pickle (without the ending 's')

````python
filename = MYDRIVE + 'pickled_file.pkl'
print("Saving file: ", filename)
pickled_file = open(filename, 'wb')
pickle.dump(mylist, pickled_file)
pickled_file.close()
````

````python

pickled_file = open(filename, 'rb')
mylistrecovered = pickle.load(pickled_file)
pickled_file.close()
print(mylistrecovered )
````

Note that for opening the file for writing we have used the `.open()` function with the `'wb'` mode, and for reading, the `.open()` function with the `'rb'` mode. In both cases the `'b'` refers to binary mode and this means that the data will be written in the form of byte objects.

We can save/recover several variables in/from the same file:

````python
mylist = ['apple', 'peach', 'orange', 'cherry']
mydict = {
  "name": "Ana",
  "surname": "García",
  "age": 25
}

pickled_file = open(filename, 'wb')
pickle.dump(mylist, pickled_file)
pickle.dump(mydict, pickled_file)
pickled_file.close()


pickled_file = open(filename, 'rb')
datarecovered1 = pickle.load(pickled_file)
datarecovered2 = pickle.load(pickled_file)
pickled_file.close()
print(datarecovered1)
print(datarecovered2)
````

## Object Oriented Programming
The common programming languages, like C, or Python (until now), work with the **Imperative Programming Paradigm**. This means: you have your data, and the programing instructions ORDER the modification of the data. The language does not bother if the data can accept that modification, or it means someting useful.

For example, you can have an integer variable to code the age of one person. You can ask the integer to take the value of 200, a valid integer value, but it does not have a sense for age values.

Other problem, is the security of the data: When you work in big projects, the variables can be modified by orders in zones of code that they are not allowed.

Or variables can be observed by functions which are not allowed to.

In order to avoid this, the Object Oriented Programming was proposed:

The information is an object. The values of this information are stored in properties of the objects. And that properties can only be modified by specific methods associated to the objects.

The description of the objects, along with its properties and associated methods is called __*CLASSES*__. The instance of one class, with specific values, is called __*OBJECT*__.

We can instantiate as many objects of an specific class as we need.

One analogy:

There is the class __*HUMAN*__. The class __*HUMAN*__ can have several properties like *First Name*, *Surname*, *age*, *gender*, *father*, *mother*, *birth date*, *death date*, etc...

An instance of a class __*HUMAN*__, could be *CHARLES*, Name: *Charles*, Surname: *Windsor*, age: *>70*, gender: *male*, mother: *Elizabeth*, father: *Phillip*.

And so on, for each one of the humans.

In class __*HUMAN*__ you can define several methods which modify or return these properties. For example, The __*HUMAN*__, can be assigned a Name, change his name, change his civil status, etc. Those changes should only be done by the authorized methods, to avoid any kind of conflict.

Now, let's write the class __*HUMAN*__ and start to write his properties and methods.

Note: To properly define its `birth_date`` we can use the *datetime* package: https://docs.python.org/3/library/datetime.html


````python
from datetime import datetime

class HUMAN:
  # Here we will write the properties:
  # None value means it is not defined
  birth_date=None
  FirstName=None
  Surname=None

  # The first method is called the constructor, it is called when we create the object

  def __init__ (self, FirstName, Surname):
    # In objects, the first parameter must always be *self*, and will be used to refer this object
    self.FirstName = FirstName
    self.Surname = Surname

  def set_birth_date (self, year, month, day):
    self.birth_date=datetime(year,month,day)

  def get_birth_date (self):
    return self.birth_date

  def get_age(self):
    now = datetime.today()
    age=now-self.birth_date
    return age.days//365

  def get_fullname(self):
    return "{0} {1}".format(self.FirstName, self.Surname)
````

````python
Trump=HUMAN("Donald", "Trump")
Trump.set_birth_date(1946, 6, 14)
print("Trump's age: {0} years".format(Trump.get_age()))
print("Trump's full name: {0}".format(Trump.get_fullname()))
````

````python
class HUMAN:
  # Here we will write the properties:
  # None value means it is not defined
  birth_date=None
  FirstName=None
  Surname=None
  civil_status = None
  children = []

  # The first method is called the constructor, it is called when we create the object
  def __init__ (self, FirstName, Surname):
    # In objects, the first parameter must always be *self*, and will be used to refer this object
    self.FirstName = FirstName
    self.Surname = Surname
    self.civil_status = "single" # Default civil status
    self.children = [] # Initialize children as an empty list

  def set_birth_date (self, year, month, day):
    self.birth_date=datetime(year,month,day)

  def get_birth_date (self):
    return self.birth_date

  def get_age(self):
    now = datetime.today()
    age=now-self.birth_date
    return age.days//365

  def get_fullname(self):
    return "{0} {1}".format(self.FirstName, self.Surname)

  def get_civil_status(self):
    return self.civil_status

  def add_child(self, child_name):
    self.children.append(child_name)

  def get_num_children(self):
    return len(self.children)

  def get_children_names(self):
    return ', '.join(self.children)

  def change_civil_status(self, new_status):
    self.civil_status = new_status
````

## Inheritance

One of the more interesant feature with objects is the __inheritance__. This allows create new classes which inherits from father classes their properties and methods an extends them with new ones:
https://www.w3schools.com/python/python_inheritance.asp

````python
class PROFESSIONAL(HUMAN):

  studies=None

  def __init__(self, FirstName, Surname, studies):
    HUMAN.__init__(self, FirstName, Surname)
    self.studies = studies

  def get_studies(self):
    return self.studies
````

````python
Trump=PROFESSIONAL("Donal", "Trump", "President")

print("Trump's profession: {0}".format(Trump.get_studies()))
````
