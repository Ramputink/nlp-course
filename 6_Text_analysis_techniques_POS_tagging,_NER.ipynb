# 2 - Part of Speech (POS) Tagging

El objetivo es asignar a cada palabra de un determinado texto una etiqueta  específica (**categoría léxica**) en función de su **contexto sintáctico y su función**.  Este etiquetado es realmente útil en NLP para obtener características específicas de un texto. Por ejemplo, podemos filtrar los sustantivos y determinar los más destacados.

En idioma inglés, utilizaremos la notación [**Penn Treebank**](https://repository.upenn.edu/cgi/viewcontent.cgi?article=1603&context=cis_reports) para el etiquetado. En este [enlace](https://repository.upenn.edu/cgi/viewcontent.cgi?article=1603&context=cis_reports) puede encontrarse una descripción detallada de cada categoría, con ejemplos.

<center>
<img src="http://www.tsc.uc3m.es/~navia/figures/tags.png" width="40%">
</center>

Un aspecto común a cualquier algoritmo de POS tagging es la construcción de un método para **estimar la etiqueta $y_i$ o categoría**  asociada a la palabra i-ésima $\mathbf{x}_i$ dadas las palabras anteriores $\mathbf{x}_j$ para $j < i$ o incluso las posteriores dentro de una ventana de procesado. Ejemplos clásicos de métodos para POS tagging son los **Modelos Ocultos de Markov, (ing. "Hidden Markov Models", HMMs)** y **Conditional Random Fields, CRFs**. No entraremos en mayor detalle en el funcionamiento de estos modelos, disponéis de más información en este [capítulo de libro](https://web.stanford.edu/~jurafsky/slp3/8.pdf).

En la actualidad, **los métodos con mayor precisión** se basan en [redes **neuronales recurrentes (RNNs)**](https://arxiv.org/abs/1510.06168) o de tipo [**transformer**](https://github.com/soutsios/pos-tagger-bert). Estos modelos de procesado de lenguaje se utilizan también en los "**Large Language Models**" (**LLMs**), fundamento de los esquemas de **Inteligencia Artificial** (**IA**) más recientes, y los revisaremos con mayor detalle más adelante.

### 2.1 - POS tagging usando spaCy


```python
# Descargamos el modelo Spacy
!python -m spacy download en_core_web_md
```

```python
# Analizamos una frase con Spacy

nlp = spacy.load("en_core_web_md")

sentence = 'I am an engineer. Some day I hope to engineer an electric engine. John builds a circuit.'

doc_spacy = nlp(sentence)
print(type(doc_spacy))
print(doc_spacy)
```

Recordemos que `doc` es un objeto iterable, compuesto por elementos tipo [`token`](https://spacy.io/api/token) que a su vez son objetos con propiedades en Spacy que nos permiten acceder a una serie de atributos léxicos, los principales son:

- **lemma**: lema de la palabra
- **pos**: la etiqueta POS simple ("PRON", "NOUN", "VERB", etc.)
- **tag**: la etiqueta POS detallada ("PRP", "NNP", "VBP", etc.)
- **dep**: el tipo de dependencia sintáctica ("nominal subject", "attribute", "noun phrase as adverbial modifier")

```python
for token in doc_spacy:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_)
    print('-' * 50)
```

# 3 - Shallow Parsing o Chunking (análisis sintáctico superficial o fragmentación)

El análisis sintáctico superficial o fragmentación (**shallow parsing (SP) or chunking**) es una técnica para **agrupar tokens adyacentes en "frases" en función de sus etiquetas POS**. En lo siguiente, entendemos una **"frase" como una agrupación de palabras sin significado completo, al carecer de estructura de sujeto-predicado**.

**SP** típicamente se centra en la identificación de **determinados tipos de frases**, tales como frases sustantivas o verbales. **SP no realiza un análisis profundo** de dichas frases para determinar la relación entre éstas mediante mediante reglas sintácticas o gramáticas. En esto último se centran métodos tales como **Constituency Parsing (CP)** y **Dependency Parsing (DP)**.



Hay **cinco categorías principales de frases**, que se describen a continuación:

- **Frase sustantiva** ([Noun Phrase](https://dictionary.cambridge.org/es/gramatica/gramatica-britanica/noun-phrases?q=Noun+phrases)  **NP**): Un sustantivo actúa como palabra principal. Las frases sustantivas actúan como sujeto u objeto directo que recibe la acción del verbo. Pueden ser reemplazadas por un pronombre sin hacer que la oración o cláusula sea sintácticamente incorrecta. Ej: "**`Mi mejor amigo de la infancia`** vive en Londres"; "**`Él`** vive en Londres".

- **Frase verbal** ([Verb phrase](https://dictionary.cambridge.org/es/gramatica/gramatica-britanica/verb-phrases?q=Verb+phrases)  **VP**): Tienen un verbo que actúa como palabra principal, incluyendo el verbo principal, infinitivos y participios. Este verbo puede ir seguido de otras frases o complementos. Las frases verbales son útiles para **comprender las acciones** en las que participan los sustantivos. Ej: "Los estudiantes **`están estudiando para el examen final`**."

- **Frase adjetiva** ([Adjective phrase](https://dictionary.cambridge.org/es/gramatica/gramatica-britanica/adjective-phrases)  **ADJP**): Son frases cuya palabra principal es un adjetivo. Su función principal es **describir o calificar sustantivos y pronombres** en una oración y se colocarán antes o después del sustantivo o pronombre. Ej: "Esa película fue **`increíblemente aburrida`**."

- **Frase de adverbio** ([Adverb phrase](https://dictionary.cambridge.org/es/gramatica/gramatica-britanica/adverb-phrases?q=Adverb+phrases) o **ADVP**): un adverbio actúa como la palabra principal de la frase. Las frases de adverbio se utilizan como **modificadores de sustantivos, verbos o adverbios** al proporcionar más detalles para describirlos o calificarlos. Ej: "El guepardo corre **`muy rápidamente`**."

- **Prepositional phrase** ([Prepositional phrase](https://dictionary.cambridge.org/es/gramatica/gramatica-britanica/prepositional-phrases?q=Prepositional+phrases) **PP**): Estas frases suelen contener una **preposición como palabra principal** y otros componentes léxicos como sustantivos, pronombres, etc. Actúan como un adjetivo o adverbio que describe otras palabras o frases. Ej: "Dejé las llaves **`sobre la mesa de cocina`**."



La detección de frases sustantivas (**NP chunking**) es la variante más común de SP. Vamos a ver cómo Spacy nos da una solución directa usando el objeto `doc.noun_chunks`.


```python
sentence = "Artificial intelligence in the context of computer science, is a discipline and a set of cognitive and intellectual capabilities expressed by computer systems or combinations of algorithms whose purpose is the creation of machines that imitate human intelligence to perform tasks, and that can improve as they collect more and more information."

doc = nlp(sentence)

for chunk in doc.noun_chunks:
    print(chunk.text)
    print('-' * 20)
```

Por otro lado, Spacy **no proporciona** una solución directa para encontrar el resto de tipos de frases. Para encontrar dichas frases hay que utilizar un **matcher de expresiones, indicando explícitamente los patrones que se están buscando**. Veremos a continuación algunos ejemplos de uso de la utilidad **`matcher`** de Spacy.


Como el proceso **POS tagging** se basa en un modelo entrenado con ejemplos, **no es infalible**. En todo caso, siempre podemos refinar el proceso usando **expresiones regulares** y la clase [**Matcher**](https://spacy.io/api/matcher) de Spacy. Se pueden encontrar muchos ejemplos de uso ['aquí'](https://spacy.io/usage/rule-based-matching#matcher). Por ejemplo, nos puede interesar encontrar las frases en las que se menciona la compra de un producto o el viaje a algún país, por ejemplo.

Esta son algunos de los atributos de **matcher** que podremos a utilizar:

- **TEXT**: el texto exacto de un token.
- **LENGTH**: La longitud del texto del token.
- **IS_ALPHA, IS_ASCII, IS_DIGIT**: el texto del token consta de caracteres alfabéticos, caracteres ASCII y dígitos.
- **IS_LOWER, IS_UPPER, IS_TITLE**: el texto del token está en minúsculas, mayúsculas y título.
- **IS_PUNCT, IS_SPACE, IS_STOP**: el token es puntuación, espacios en blanco y palabra vacía.
- **IS_SENT_START**: el token es el comienzo de la oración.
- **LIKE_NUM, LIKE_URL, LIKE_EMAIL**: el texto del token se parece a un número, URL o correo electrónico.
- **POS, TAG, MORPH, DEP, LEMMA, SHAPE**: la etiqueta de parte del discurso simple y extendida del token, análisis morfológico, etiqueta de dependencia, lema, forma.
- **ENT_TYPE**: tipo de entidad NER (lo veremos a continuación).
- **REGEX**: expresión regular.

```python
id = "carroll-alice.txt"
book = gutenberg.raw(id)

print("El libro {} contiene {} caracteres.".format(id, len(book)))

print("\nEste es el inicio del libro:\n")
print(book[:500] + " (...)")

# A efectos del procesado de texto, vamos a eliminar los retornos:
book = book.replace('\n', ' ')

print("\nEste es el inicio del libro, sin retornos:\n")
print(book[:500] + " (...)")
```

Vamos a utilizar la herramienta "**matcher**" para encontrar **patrones específicos**. Comenzaremos por localizar verbos:

```python
# Activamos el matcher y procesamos el libro con Spacy
from spacy.matcher import Matcher
book_spacy = nlp(book)
```

```python
# Buscamos verbos en el vocabulario del libro:
patron1 = [{"POS": "VERB"}]
buscador = Matcher(nlp.vocab)
buscador.add("Patron1", [patron1])
coincidencias = buscador(book_spacy)
print(coincidencias)
```

La herramienta **`matcher`** devuelve una lista de coincidencias. Cada coincidencia tiene tres valores:

- un identificador de la búsqueda,

- la posición inicio en la lista de tokens

- la posición final


```python
# Presentamos resultados
print("La respuesta de matcher es una lista de coincidencias")
print(coincidencias)
print("\nSe han encontrado {} coincidencias con el patrón de búsqueda.".format(len(coincidencias)))

print("\nEstos son los 10 primeros patrones coincidentes encontrados (verbos):\n")
for caso in coincidencias[:10]:
    print (book_spacy[caso[1]:caso[2]])
```

```python
# Otro ejemplo de patrón de búsqueda
# Buscamos verbos de 6 caracteres:
patron1 = [{"POS": "VERB", "LENGTH": 6}]

# Si queremos resetear el patrón de búsqueda, es importante reiniciar el matcher:
buscador = Matcher(nlp.vocab)
buscador.add("Patron1", [patron1])
coincidencias = buscador(book_spacy)

print("\nEstos son los 10 primeros patrones coincidentes encontrados (verbos de 6 caracteres):\n")
for caso in coincidencias[:10]:
    print (book_spacy[caso[1]:caso[2]])
```

```python
# Buscamos verbo seguido de palabra de 6 caracteres, observar diferencia con búsqueda anterior:
patron1 = [{"POS": "VERB"}, {"LENGTH": 6}]
buscador = Matcher(nlp.vocab)
buscador.add("Patron1", [patron1])
coincidencias = buscador(book_spacy)

print("\nEstos son los 10 primeros patrones coincidentes encontrados:\n")
for caso in coincidencias[:10]:
    print (book_spacy[caso[1]:caso[2]])
```

```python
# Hacemos un poco más específica la búsqueda:
verbos_a_identificar = ["think"]
patron1 = [{"POS": "VERB", "LEMMA": {"IN": verbos_a_identificar}}]

buscador = Matcher(nlp.vocab)
buscador.add("Patron1", [patron1])
coincidencias = buscador(book_spacy)

print("\nSe han encontrado {} coincidencias con el patrón de búsqueda.".format(len(coincidencias)))
print("\nEstos son los 10 primeros patrones coincidentes encontrados (mostramos dos palabras antes y después)\n:")
for caso in coincidencias[:10]:
    print (book_spacy[caso[1]-2:caso[1]], "[", book_spacy[caso[1]:caso[2]] , "]", book_spacy[caso[2]:caso[2]+2])
```

# 4 - Reconocimiento de entidades (Name entity recognition NER)

Si bien **POS tagging** nos indicará que palabras como `Janet`, `Stanford University` o `Colorado` son **nombres propios**, estos nombres propios se refieren a **tipos de entidades distintas**: `Janet` es una **persona**, `Stanford University` es una **organización** y `Colorado` es una **entidad geopolítica**.

La tarea de **reconocimiento de entidades (ing. "Name Entity Recognition", NER)** consiste tanto en encontrar tramos de texto que constituyan **nombres propios** como etiquetar el **tipo de entidad**. Las cuatro etiquetas de entidad son las más comunes: **PER (persona)**, **LOC (ubicación)**, **ORG (organización)** o **GPE (entidad geopolítica)**. Sin embargo, el término entidad se extiende comúnmente para incluir cosas que no son entidades per se, incluidas fechas, horas y otros tipos de expresiones temporales, e incluso expresiones numéricas como precios.

Los distintos modelos NER son fundamentalmente **algoritmos de aprendizaje automático entrenados sobre bases de datos etiquetadas**. Por tanto, su rendimiento está ligado al tipo de texto en que fueron entrenados y esto implica que para un nuevo texto pueden no alcanzar las mismas prestaciones.  Si se considera necesario, [Spacy permite hacer un ajuste fino](https://course.spacy.io/en/chapter4) (reentrenar) cualquiera de sus modelos a un nuevo corpus de documentos.

# 5 - Dependency parsing

**Dependency parsing** es el proceso de extraer el **análisis sintáctico de dependencias de una oración** para representar su **estructura gramatical**. Define la relación de dependencia entre las palabras clave y sus dependientes:

- El **encabezado** de una frase no tiene dependencia y se llama raíz (**ROOT**). Suele ser un verbo.
- El **resto de palabras** están **vinculadas al encabezado**.

Este análisis proporciona información sobre **qué papel juega una palabra en el texto y cómo se relacionan las diferentes palabras entre sí**. También se utiliza en el *shallow parsing* y en NER.

Las dependencias se codifican en una **estructura de tipo árbol**: las palabras son los nodos que están conectadas en función de su dependencia gramatical.





```python
sentence = 'I read a book'
doc_spacy = nlp(sentence)
NER_data = []
for token in doc_spacy:
  NER_data.append((token.text, token.tag_, token.head.text, token.dep_))

df = pd.DataFrame(NER_data, columns=['Palabra', 'POS TAG', 'HEAD', 'DEP'])
df.style.set_properties(**{'text-align': 'left'})
display(df)
```

En el ejemplo anterior, podemos ver **distintos tipos de relaciones entre las palabras**. Una lista completa puede consultarse en este enlace sobre [**Stanford typed dependencies**](https://nlp.stanford.edu/software/dependencies_manual.pdf). Las más habituales son:

- **Sujeto**, nsubj (nominal subject): El sujeto de la oración.

- **Objeto directo**, dobj (direct object): El objeto directo sobre el que recae la acción.

- **Objeto indirecto**, iobj (indirect object): El objeto indirecto.

- **Modificadores de Sustantivos**: amod (adjectival modifier), det (determiner), nmod (nominal modifier)

- **Modificadores de Verbos y Adverbios**: advmod (adverbial modifier), aux (auxiliary verb)

- **Relaciones de Estructura y Otros**, root (root): La palabra principal de la oración (usualmente el verbo principal), cc (coordinating conjunction), conj (conjunct), punct (punctuation).

## Solución:

```python
from spacy.matcher import Matcher
#all_books_spacy = nlp(all_books)
parte_books = all_books[0:150000]
books_spacy = nlp(parte_books)

patron1 = [{"ENT_TYPE": "PERSON"}]
patron2 = [{"ENT_TYPE": "PERSON"}, {"ENT_TYPE": "PERSON"}]
patron3 = [{"ENT_TYPE": "PERSON"}, {"ENT_TYPE": "PERSON"}, {"ENT_TYPE": "PERSON"}]

buscador = Matcher(nlp.vocab)
buscador.add("Patron1", [patron1])
buscador.add("Patron2", [patron2])
buscador.add("Patron3", [patron3])
coincidencias = buscador(books_spacy)

names = []
for caso in coincidencias:
    names.append(all_books_spacy[caso[1]:caso[2]].text)

names_L1 = [name for name in names if len(name.split()) == 1]
names_L2 = [name for name in names if len(name.split()) == 2]
names_L3 = [name for name in names if len(name.split()) == 3]

# Filtramos los nonbres sencillos que aparecen en los dobles
nombres_multiples = " ".join(names_L2) + " " + " ".join(names_L3)

names_L1_filtrados = [name for name in names_L1 if name not in nombres_multiples]

names = names_L1_filtrados + names_L2 + names_L3

names = list(set(names))

# Filtramos los nombres:

names_OK = []
names_deleted = []

for name in names:
  remove = False
  for token in nlp(name):
    if not token.is_alpha:
      remove = True
    if token.is_punct:
      remove = True
    if token.text.lower() == token.text:
      remove = True

  if not remove:
    names_OK.append(name)
  else:
    names_deleted.append(name)

names_OK.sort()

print("\nEsta es la lista de personajes filtrada, en orden alfabético:\n")
for name in names_OK:
  print(name)

print("\nEsta es la lista de nombres eliminados:\n")
for name in names_deleted:
  print(name)
```
